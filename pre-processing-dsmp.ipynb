```python
import pandas as pd

# Filter out rows with 'complex.id' equal to 0
#df_filtered = df[df['complex.id'] != 0]

# Drop the 'gene' column since it's redundant after this point
cdr3_alpha_df = cdr3_alpha_df.drop('gene', axis=1)
cdr3_beta_df = cdr3_beta_df.drop('gene', axis=1)

# Rename columns to include the chain type
cdr3_alpha_df.rename(columns=lambda x: x if x in ['complex.id', 'species', 'mhc.a', 'mhc.b', 'mhc.class', 'antigen.epitope']
                else x + '_alpha', inplace=True)
cdr3_beta_df.rename(columns=lambda x: x if x in ['complex.id', 'species', 'mhc.a', 'mhc.b', 'mhc.class', 'antigen.epitope']
               else x + '_beta', inplace=True)
```

```python
cdr3_alpha_df = cdr3_alpha_df[cdr3_alpha_df['complex.id'] != 0]
cdr3_beta_df = cdr3_beta_df[cdr3_beta_df['complex.id'] != 0]
```

```python
cdr3_alpha_df, cdr3_beta_df
```

```python
# Merge on 'complex.id' and other identical columns, using suffixes for the differing ones
cdr3_alpha_beta_df = pd.merge(
    cdr3_alpha_df, 
    cdr3_beta_df, 
    on=['complex.id', 'species', 'mhc.a', 'mhc.b', 'mhc.class', 'antigen.epitope'],
    suffixes=('_alpha', '_beta')
)

# Assuming 'cdr3', 'v.segm', and 'j.segm' are the only columns that need suffixes
# Now, we have columns like 'cdr3_alpha', 'v.segm_alpha', 'j.segm_alpha' and their '_beta' counterparts


# Output the head of the DataFrame to ensure it's merged as expected
print(cdr3_alpha_beta_df.head())
```
