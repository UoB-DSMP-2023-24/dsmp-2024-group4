Performing different types of encoding for the 'cdr3' column

# One-Hot Encoding

```python
import pandas as pd
import numpy as np


# Define the one-hot encoding function
def one_hot_encode_sequence(sequence, amino_acids):
    # Create an array for storing the one-hot encoded sequence
    encoding = np.zeros((len(sequence), len(amino_acids)), dtype=int)
    for i, aa in enumerate(sequence):
        if aa in amino_acids:
            aa_index = amino_acids.index(aa)
            encoding[i, aa_index] = 1
    # Flatten the array to a 1D array
    return encoding.flatten()

# List of all amino acids for the one-hot encoding
amino_acids = 'ACDEFGHIKLMNPQRSTVWY'

# Apply the one-hot encoding to the 'cdr3' column
cdr3_one_hot_encoded = filtered_data['cdr3'].apply(lambda seq: one_hot_encode_sequence(seq, amino_acids))

# Now you have a new column 'cdr3_one_hot_encoded' in your DataFrame with the one-hot encoded sequences
print(cdr3_one_hot_encoded)
```

# BLOSUM 62 ENCODING

```python
import numpy as np
from Bio.Align import substitution_matrices

# Load the BLOSUM62 matrix
blosum62 = substitution_matrices.load("BLOSUM62")

# Define a function to encode a single amino acid sequence
def encode_sequence_with_blosum62(sequence, blosum62_matrix):
    # Initialize an empty list to store the encoded sequence
    encoded_sequence = []
    # Iterate over each amino acid in the sequence
    for aa in sequence:
        # For each amino acid, retrieve its BLOSUM62 encoding
        # If the amino acid is not in the matrix (e.g., a gap or special character), we can use a vector of zeros
        aa_encoding = blosum62_matrix.get(aa, np.zeros(len(blosum62_matrix)))
        encoded_sequence.append(aa_encoding)
    # Convert the list of encoded amino acids into a numpy array
    return np.array(encoded_sequence)

# Example: Encoding the 'cdr3' sequence of the first row
sequence_cdr3 = filtered_data.loc[0, 'cdr3']
cdr3_blosum_encoded = encode_sequence_with_blosum62(sequence_cdr3, blosum62)
print(cdr3_blosum_encoded)
```

```python
import pandas as pd
import numpy as np
from Bio.Align import substitution_matrices

# Load your DataFrame
# df = pd.read_csv('path_to_your_data.csv')

# Load the BLOSUM62 matrix
blosum62 = substitution_matrices.load("BLOSUM62")

# Define the encoding function
def encode_sequence_with_blosum62(sequence, blosum62_matrix):
    # Initialize an empty list to store the encoded sequence
    encoded_sequence = []
    # Iterate over each amino acid in the sequence
    for aa in sequence:
        # If the amino acid is in the BLOSUM62 matrix, retrieve its encoding
        # If not (e.g., a gap or special character), use a vector of zeros
        encoded_sequence.append(blosum62_matrix.get(aa, np.zeros((1, len(blosum62_matrix)))))
    # Convert the list of encoded amino acids into a numpy array
    return np.concatenate(encoded_sequence)

# Apply the function to each cdr3 sequence in the DataFrame
cdr3_encoded = filtered_data['cdr3'].apply(lambda x: encode_sequence_with_blosum62(x, blosum62))

# Each row in 'cdr3_blosum62_encoded' will now be a flat array of BLOSUM62 scores
print(cdr3_encoded)
```
