```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#txt path
txt_path = 'vdjdb-2023-06-01/vdjdb.txt'

#read txt file
df = pd.read_csv(txt_path, sep='\t')

#show the first 5 rows without hiding any columns
df.head()
```

```python
#将txt文件转换为csv文件
df.to_csv('vdjdb.csv', index=False)
```

```python
# Analyze the data for each column
df.info()
df.describe(include='all')

# Check for missing values
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Creating subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 15))


# Plot for gene frequency
sns.countplot(x='gene', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Frequency of gene types')

# Plot for v.segm frequency
sns.countplot(x='v.segm', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Frequency of v.segm types')
axes[0, 1].tick_params(axis='x', rotation=90)
axes[0, 1].set(xticklabels=[])
axes[0, 1].set(xlabel=None)

# Plot for j.segm frequency
sns.countplot(x='j.segm', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Frequency of j.segm types')
axes[1, 0].tick_params(axis='x', rotation=90)
axes[1, 0].set(xticklabels=[])
axes[1, 0].set(xlabel=None)

# Plot for species frequency
sns.countplot(x='species', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Frequency of species')



plt.tight_layout()
plt.show()
```
